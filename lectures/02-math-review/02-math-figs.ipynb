## Calculus

```python
x = np.linspace(-5, 5, 100)
y = x**2 + 5
plt.plot(x, y, label="$y = x^2 + 5$")
dy_dx = np.gradient(y, x)
plt.plot(x, dy_dx, label=r"$\frac{dy}{dx} = 2x$")
plt.annotate(r"$\frac{dy}{dx}$ = 0", (0,5), xytext=(-1, 10), arrowprops={"arrowstyle": "->"})
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 11, 100)
y1 = np.sin(x)
y2 = np.abs(x)

plt.plot(x, y1, label="$y = sin(x)$")
plt.plot(x, y2, label="$y = |x|$")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig("../figures/02-differentiable.png")
```

## Vector Calculus

```python
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = 2 * X1 + X2 + X1 * X2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X1, X2, Y, alpha=0.8, cmap="viridis")
fig.colorbar(surf)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("$y = 2x_1 + x_2 + x_1x_2$")
plt.savefig("../figures/02-surface.png")
```

```python
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma,100)
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
plt.plot(x, y, label="PDF")
plt.plot(x, (x[1] - x[0]) * np.cumsum(y), label="CDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Normal Distribution")
plt.savefig("../figures/02-normal_dist.png")
plt.show()
```
